```python
import pandas as pd
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
data.loc[data['whoAmI'] == 'robot', 'robot'] = "1"
data.loc[data['whoAmI'] == 'human', 'human'] = "1"
data.fillna(0, inplace=True)
```
